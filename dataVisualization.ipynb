```python
import matplotlib.pyplot as plt
import pandas as pd

df = pd.read_csv("SolarPrediction.csv")
# Assuming 'df' is your DataFrame containing the data

# Convert 'Data' and 'Time' columns to datetime format
df['DateTime'] = pd.to_datetime(df['Date'] + ' ' + df['Time'])

# Set 'DateTime' as the index
df.set_index('DateTime', inplace=True)

# Plotting
plt.figure(figsize=(14, 8))

# Temperature
plt.subplot(3, 2, 1)
plt.plot(df.index, df['Temperature'], color='blue')
plt.title('Temperature')
plt.xlabel('Date and Time')
plt.ylabel('Temperature (°C)')

# Humidity
plt.subplot(3, 2, 2)
plt.plot(df.index, df['Humidity'], color='green')
plt.title('Humidity')
plt.xlabel('Date and Time')
plt.ylabel('Humidity (%)')

# Pressure
plt.subplot(3, 2, 3)
plt.plot(df.index, df['Pressure'], color='red')
plt.title('Pressure')
plt.xlabel('Date and Time')
plt.ylabel('Pressure (hPa)')

# Wind Direction
plt.subplot(3, 2, 4)
plt.plot(df.index, df['WindDirection(Degrees)'], color='orange')
plt.title('Wind Direction')
plt.xlabel('Date and Time')
plt.ylabel('Wind Direction (Degrees)')

# Wind Speed
plt.subplot(3, 2, 5)
plt.plot(df.index, df['Speed'], color='purple')
plt.title('Wind Speed')
plt.xlabel('Date and Time')
plt.ylabel('Wind Speed (m/s)')

# Global Irradiance
plt.subplot(3, 2, 6)
plt.plot(df.index, df['Radiation'], color='brown')
plt.title('Global Irradiance')
plt.xlabel('Date and Time')
plt.ylabel('Global Irradiance')

# Adjust layout
plt.tight_layout()

# Show plot
plt.show()
```
